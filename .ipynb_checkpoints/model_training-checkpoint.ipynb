```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
try:
    X_train = pd.read_csv('X_train.csv')
    X_test = pd.read_csv('X_test.csv')
    y_train = pd.read_csv('y_train.csv').values.ravel() # .ravel() fixes the "column-vector" warning
    y_test = pd.read_csv('y_test.csv').values.ravel()
except FileNotFoundError:
    print('Error in loading the files')
    exit()
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(max_depth=5),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
}
```

```python
best_model_name = ''
best_accuracy = 0
best_model_object = None
```

```python
for name, model in models.items():
    print(f'Training {name}')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(classification_report(y_test, y_pred))
    print(f'Accuracy: {acc*100:.2f}%')

    if acc > best_accuracy:
        best_accuracy = acc
        best_model_name = name
        best_model_object = model
```

```python
print(f"Best model is: {best_model_name} with {best_accuracy*100:.2f}% Accuracy!")
```

```python
print(f"\n--- Detailed Report for {best_model_name} ---")
y_pred_final = best_model_object.predict(X_test)
print(classification_report(y_test, y_pred_final))
```

```python
cm = confusion_matrix(y_test, y_pred_final)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title(f'Confusion Matrix ({best_model_name})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('confusion_matrix.png')
```

```python
with open('model.pkl', 'wb') as file:
    pickle.dump(best_model_object, file)
```
